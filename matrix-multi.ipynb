```python
import torch
torch.manual_seed(2026)
```

```python
a = torch.ones((3, 3))
b = torch.randint(0, 10, (3, 2)).float()
c = a @ b

print('a =')
print(a)
print('----------')
print('b =')
print(b)
print('----------')
print('c =')
print(c)    
```

```python
a = torch.tril(torch.ones(3, 3))
# b = torch.randint(0, 10, (3, 2)).float()
a = a / torch.sum(a, dim=1, keepdim=True) # normalize the rows to sum to 1      
c = a @ b

print('a =')
print(a)
print('----------')
print('b =')
print(b)
print('----------')
print('c =')
print(c)    
```
